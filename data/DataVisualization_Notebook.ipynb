# Week 10 Hands-On: Data Visualization

This notebook is designed for **DESN2003 Week 10**. The goal is to give students a short, editable workshop they can run in **Google Colab** without downloading extra files first.

By the end, students should be able to:
- inspect a small dataset before plotting it
- compare groups with box plots, histograms, and bar charts
- look for patterns with a heatmap and scatter plot
- create a quick qualitative visualization from transcript text
- tweak a working example instead of starting from a blank page

The sample data is **baked into the notebook**. The notebook also writes out `survey_data.csv` and `transcripts.json` in the runtime so students can inspect or reuse them.


````python
from io import StringIO
from pathlib import Path
import json

from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

try:
    from wordcloud import STOPWORDS, WordCloud
except ImportError:
    import subprocess
    import sys

    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', 'wordcloud'])
    from wordcloud import STOPWORDS, WordCloud

sns.set_theme(style='whitegrid', context='talk')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
pd.set_option('display.max_colwidth', 140)
````

````python
survey_csv = """ParticipantID,Group,Region,Score,Satisfaction,Recommend,Comment
1,GenAI,Africa,88,5,Yes,The AI tutor made studying much easier.
2,Traditional,Africa,65,3,No,Studying without AI was a bit challenging.
3,GenAI,Asia,79,4,Yes,The AI answered my questions quickly.
4,GenAI,Asia,78,4,Yes,The AI provided clear explanations.
5,Traditional,Asia,70,4,Yes,Studying on my own worked fine for me.
6,Traditional,Asia,60,3,No,"It was okay, but I think AI might have helped."
7,GenAI,Europe,82,4,Yes,Having an AI assistant was extremely helpful.
8,GenAI,Europe,85,5,Yes,Using the AI tool saved me a lot of time.
9,Traditional,Europe,68,3,No,"I did all right, but it took a lot of time."
10,Traditional,Europe,58,2,No,Learning without AI was a good experience.
11,GenAI,North America,92,5,Yes,ChatGPT was a great help in my studies.
12,GenAI,North America,81,4,Yes,Using the AI support improved my understanding.
13,GenAI,North America,66,2,No,"The AI's answers were often confusing."
14,GenAI,North America,74,3,No,The AI was somewhat helpful but not perfect.
15,Traditional,North America,75,4,Yes,I did well without using any AI.
16,Traditional,North America,77,4,Yes,I felt confident even without an AI helper.
17,Traditional,South America,50,1,No,It was really hard without any AI assistance.
18,Traditional,South America,72,3,No,I struggled a lot to learn by myself.
19,GenAI,South America,90,5,Yes,I found the AI help very useful for learning.
20,Traditional,South America,83,5,Yes,I managed to learn everything without AI.
"""

transcripts = [
    {
        'id': 1,
        'group': 'GenAI',
        'text': (
            'I tried using ChatGPT to help me study for my exams. At first I was not sure, '
            'but it turned out to be extremely helpful. Whenever I had a question, the AI '
            'provided an explanation right away. It felt like having a tutor available 24/7. '
            'It saved me so much time because I did not have to search for answers on my own. '
            'I am definitely going to keep using such tools for learning.'
        ),
    },
    {
        'id': 2,
        'group': 'GenAI',
        'text': (
            'I used an AI chatbot for help in my course, but it was not as great as I expected. '
            'Sometimes it gave me answers that were confusing or even wrong. I ended up spending '
            'extra time double-checking the information it provided. It was frustrating because I '
            'was never sure if I could trust the answers. In the end, I felt like it did not '
            'improve my learning much.'
        ),
    },
    {
        'id': 3,
        'group': 'Traditional',
        'text': (
            'I decided not to use any AI or online tutors while studying. I wanted to do '
            'everything myself using textbooks and class notes. It was a real challenge and '
            'took me a lot longer to get through the material. There were times I got stuck '
            'on a concept and had no one to ask for help. I think I learned the material '
            'eventually, but I was pretty stressed. After seeing others use AI tools, I am '
            'considering trying it in the future.'
        ),
    },
]

Path('survey_data.csv').write_text(survey_csv, encoding='utf-8')
Path('transcripts.json').write_text(json.dumps(transcripts, indent=2), encoding='utf-8')

survey_df = pd.read_csv(StringIO(survey_csv))
survey_df['CommentLength'] = survey_df['Comment'].str.split().str.len()
survey_df['PositiveRecommend'] = survey_df['Recommend'].eq('Yes').astype(int)
transcript_df = pd.DataFrame(transcripts)

survey_df.head()
````

## 1. Inspect The Data First

Before plotting anything, make sure students can answer basic questions:
- What columns do we have?
- Which columns are numeric, and which are categories?
- How many students are in each group?
- What does the text feedback actually sound like?


````python
print(f'Rows, columns: {survey_df.shape}')
print()
print('Column types:')
survey_df.info()
print()
print('Numeric summary:')
display(survey_df.describe(numeric_only=True).round(2))
print()
print('Group summary:')
display(survey_df.groupby('Group')[['Score', 'Satisfaction', 'CommentLength']].mean().round(2))
print()
print('Recommend by group:')
display(pd.crosstab(survey_df['Group'], survey_df['Recommend']))
````

## 2. Quantitative Visualizations

These examples are deliberately small and editable. Students can change one line at a time and rerun the cell.


````python
palette = {'GenAI': '#2A9D8F', 'Traditional': '#E76F51'}

plt.figure(figsize=(7, 5))
ax = sns.boxplot(
    data=survey_df,
    x='Group',
    y='Score',
    hue='Group',
    dodge=False,
    palette=palette,
)
if ax.legend_:
    ax.legend_.remove()
plt.title('Test Score Distribution by Group')
plt.xlabel('Learning Method')
plt.ylabel('Test Score')
plt.show()
````

````python
plt.figure(figsize=(8, 5))
sns.histplot(
    data=survey_df,
    x='Score',
    hue='Group',
    bins=8,
    multiple='layer',
    alpha=0.45,
    palette=palette,
)
plt.title('Score Distribution')
plt.xlabel('Test Score')
plt.ylabel('Number of Students')
plt.show()
````

````python
recommend_palette = {'Yes': '#2A9D8F', 'No': '#E76F51'}

plt.figure(figsize=(7, 5))
sns.countplot(data=survey_df, x='Group', hue='Recommend', palette=recommend_palette)
plt.title('Would Students Recommend The Method?')
plt.xlabel('Learning Method')
plt.ylabel('Number of Students')
plt.legend(title='Recommend')
plt.show()
````

````python
region_scores = (
    survey_df.groupby(['Region', 'Group'], as_index=False)['Score']
    .mean()
    .sort_values(['Region', 'Group'])
)

plt.figure(figsize=(11, 5))
sns.barplot(data=region_scores, x='Region', y='Score', hue='Group', palette=palette)
plt.title('Average Score By Region')
plt.xlabel('Region')
plt.ylabel('Average Score')
plt.xticks(rotation=20, ha='right')
plt.legend(title='Group')
plt.show()
````

````python
numeric_cols = ['Score', 'Satisfaction', 'CommentLength', 'PositiveRecommend']
corr = survey_df[numeric_cols].corr()

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.heatmap(corr, annot=True, cmap='YlGnBu', vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title('Correlation Heatmap')

sns.scatterplot(
    data=survey_df,
    x='Score',
    y='Satisfaction',
    hue='Group',
    s=120,
    palette=palette,
    ax=axes[1],
)
axes[1].set_title('Score vs Satisfaction')
axes[1].set_ylim(0.8, 5.2)

plt.tight_layout()
plt.show()
````

## 3. Qualitative Visualizations

A word cloud is fast and approachable, but it is **not** the same thing as proper qualitative coding. It works well here as a low-friction classroom exercise.


````python
transcript_df
````

````python
all_text = ' '.join(item['text'] for item in transcripts)
custom_stopwords = STOPWORDS.union({'ai', 'chatgpt', 'using', 'used', 'tool', 'tools'})

wordcloud = WordCloud(
    width=900,
    height=450,
    background_color='white',
    stopwords=custom_stopwords,
    collocations=False,
).generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud Of Transcript Text')
plt.show()
````

````python
theme_keywords = {
    'Helpful AI': ['helpful', 'tutor', 'great help', 'clear explanations', 'improved'],
    'Saves time': ['time', 'quickly', 'right away', 'saved'],
    'Trust / accuracy issues': ['confusing', 'wrong', 'double-checking', 'trust'],
    'Struggle without AI': ['without any ai', 'without ai', 'challenge', 'no one to ask', 'stressed'],
}

theme_hits = []
for item in transcripts:
    text = item['text'].lower()
    for theme, keywords in theme_keywords.items():
        if any(keyword in text for keyword in keywords):
            theme_hits.append({'TranscriptID': item['id'], 'Theme': theme})

theme_df = pd.DataFrame(theme_hits).drop_duplicates()
theme_summary = (
    theme_df['Theme']
    .value_counts()
    .rename_axis('Theme')
    .reset_index(name='Mentions')
)

plt.figure(figsize=(9, 5))
sns.barplot(data=theme_summary, x='Mentions', y='Theme', hue='Theme', dodge=False, palette='crest')
ax = plt.gca()
if ax.legend_:
    ax.legend_.remove()
plt.title('Simple Theme Count From Transcript Keywords')
plt.xlabel('Number of Transcripts Mentioning The Theme')
plt.ylabel('Theme')
plt.show()

display(theme_summary)
````
